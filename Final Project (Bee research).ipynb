## visualization of missing data 

```python
# identify and visualize missing values
# bar chart
import missingno as msno
msno.bar(df)
# the lower the bar is, the more missing values it has
```

```python
# Matrix 
msno.matrix(df)
# When there is no missing value, it is fully colored in black
```

```python
msno.heatmap(df)
'''
The heatmap helps us to identify a relationship in the presence of null values
between each of the columns
Close to -1: there is an anti-correlation between 2 columns: if column A has 
values, then column B has null values.
Close to 0: no correlation due to the presence of null values between the 
columns
Close to 1: strong correlation due to the presence of null values between the 
columns
'''
```

```python
msno.dendrogram(df)
```

```python
# Removal from string columns
for col in df.select_dtypes(include=[object]):
    df[col] = df[col].replace(r'^\s*$', np.nan, regex=True)
```

```python
ax = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=ax)

plt.title('Missing Values', fontsize=12)
plt.xlabel('Columns', fontsize = 10) 
plt.ylabel('Missing Values', fontsize = 10)
plt.show()
```
